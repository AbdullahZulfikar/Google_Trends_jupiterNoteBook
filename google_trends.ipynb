```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df_tesla= pd.read_csv('data/TESLA Search Trend vs Price.csv')
df_bitcoin= pd.read_csv('data/Daily Bitcoin Price.csv')
df_unemployment= pd.read_csv('data/UE Benefits Search vs UE Rate 2004-20.csv')

print(f"The value missing value in tesla is {df_tesla.isna().values.any()}")
print(f"The value missing value in bitcoin is {df_tesla.isna().values.any()}")
print(f"The value missing value in unemployment is {df_tesla.isna().values.any()}")

df_tesla.MONTH = pd.to_datetime(df_tesla.MONTH)
df_bitcoin.DATE = pd.to_datetime(df_bitcoin.DATE)
df_bitcoin = df_bitcoin.resample('M',on="DATE").last()


# this is the pyplot
plt.figure(figsize=(14,8), dpi=120) 
plt.title('Tesla Web Search vs Price', fontsize=18)



## dealing with the dates and locators
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')


# Also, increase fontsize and linewidth for larger charts
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.set_ylabel('TSLA Stock Price', color='#E6232E', fontsize=14)
ax2.set_ylabel('Search Trend', color='skyblue', fontsize=14)
## this is for grid =====
ax1.grid(color='grey',linestyle='--')

##=======================
ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE, color='#E6232E', linewidth=3,linestyle='--')
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, color='skyblue', linewidth=3,marker='o')
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)
# Displays chart explicitly
plt.show()


```
